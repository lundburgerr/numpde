```python
# a)-b)
import numpy as np

def midpoint(f,a,b):
    return (f(0.5*(a+b))*(b-a))

def trapezoid(f,a,b):
    return ((f(a)+f(b))*0.5*(b-a))

def simpson(f,a,b):
    m=0.5*(a+b)
    return ((f(a)+4*f(m)+f(b))*(b-a)/6.0)

a=0.0; b=1.0

M=5 # number of monomials to integrate
T=np.zeros((M,3+1))

for i in range(1,M+1):
    f=(lambda x: x**i)
    ex=1.0/(i+1.0)
    T[i-1]=[i,abs(midpoint (f,a,b)-ex),abs(trapezoid(f,a,b)-ex),abs(simpson(f,a,b)-ex)]
print(T)
```
